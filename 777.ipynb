```python
import os
import shutil
import librosa

input_folder = '../All_Data/Data/LJSpeech/wavs'
output_folder = '../All_Data/Data/LJSpeech/wav_le10'
output_file = 'LJSpeech_text_le10.txt'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

with open(output_file, 'w') as f:
    for filename in os.listdir(input_folder):
        if filename.endswith('.wav'):
            file_path = os.path.join(input_folder, filename)
            duration = librosa.get_duration(filename=file_path)
            if duration <= 10:
                shutil.copy(file_path, output_folder)
                f.write(filename + '\n')
```

```python
import os
import wave
import numpy as np
import matplotlib.pyplot as plt

# Define the folder containing the WAV files
folder_path = "../All_Data/Data/NEB/wavs"

# Define a function to calculate the duration of a WAV file
def get_wav_duration(file_path):
    with wave.open(file_path, 'r') as wav_file:
        num_frames = wav_file.getnframes()
        frame_rate = wav_file.getframerate()
        duration = num_frames / float(frame_rate)
        return duration

# Initialize a list to store the durations of all the WAV files in the folder
durations = []

# Loop through all the WAV files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith(".wav"):
        file_path = os.path.join(folder_path, file_name)
        duration = get_wav_duration(file_path)
        durations.append(duration)

# Calculate the mean and standard deviation of the durations
mean_duration = np.mean(durations)
std_duration = np.std(durations)

# Create a histogram of the durations
plt.hist(durations, bins=20)

# Add labels to the histogram
plt.xlabel("Duration (seconds)")
plt.ylabel("Number of WAV files")
plt.title("Distribution of WAV file durations")

# Display the histogram
plt.show()

# Print the mean and standard deviation of the durations
print("Mean duration: {:.2f} seconds".format(mean_duration))
print("Standard deviation: {:.2f} seconds".format(std_duration))
```
